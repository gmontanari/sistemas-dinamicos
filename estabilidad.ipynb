# Cuasiperiodicidad

con b = 1 / pi

```python
def f(x, b):
    return (x + b)%1
y = np.arange(0,1,0.01)
x = [0.4,]
x1 = [0.4,]
b = 1/np.pi

for t in range(len(y)-1):
    x.append(f(x[t], b))
    x1.append(f(f(x[t], b), b))
fig = plt.plot(y,y, color="orange")
fig = plt.plot(y,x)
fig = plt.plot(y, x1)
```

## con b = sqrt(2)

```python
x = [0.4,]
x1 = [0.4,]
b = np.sqrt(2)

for t in range(len(y)-1):
    x.append(f(x[t], b))
    x1.append(f(f(x[t], b), b))
fig = plt.plot(y,y, color="orange")
fig = plt.plot(y,x)
fig = plt.plot(y, x1)
```

## con b=e

```python
x = [0.4,]
x1 = [0.4,]
b = np.e

for t in range(len(y)-1):
    x.append(f(x[t], b))
    x1.append(f(f(x[t], b), b))
fig = plt.plot(y,y, color="orange")
fig = plt.plot(y,x)
fig = plt.plot(y, x1)
```

## poblaci√≥n de moscas

```python
def f(x):
    r=0.25
    K=15
    if x < K:
        return 0.01 * (x**2)
    elif x >= K:
        return 0.01*(K**2)*np.exp(-r*(x-K))

        
x = [9,]
for t in range(30):
    x.append(f(x[t]))
fig=plt.plot(x)
```
