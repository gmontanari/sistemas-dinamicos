```python
# importamos bibliotecas para plotear
import matplotlib
import matplotlib.pyplot as plt

# para desplegar los plots en el notebook
%matplotlib inline

# ejecutarlas diez mil veces, medir su tiempo
scualos=[11.9,
        21.4,
        22.1,
        21.2,
        36.4,
        27.3,
        16.0,
        15.9,
        14.8,
        10.7]

# armar gráfica de barras
plt.bar(range(1914,1924), scualos, align='center', alpha=0.5)
#plt.title('10,000 ejecuciones, k=25')
plt.ylabel('% scualos')
plt.xticks(range(1914,1924),[str(n) for n in range(1914,1924)])
plt.show()
```

```python
# importamos bibliotecas para plotear
import matplotlib
import matplotlib.pyplot as plt

# para desplegar los plots en el notebook
%matplotlib inline

# ejecutarlas diez mil veces, medir su tiempo
scualos=[11.9,
        21.4,
        22.1,
        21.2,
        36.4,
        27.3,
        16.0,
        15.9,
        14.8,
        10.7]

# armar gráfica de barras
plt.bar(range(1914,1924), [100-n for n in scualos], align='center', alpha=0.5)
#plt.title('10,000 ejecuciones, k=25')
plt.ylabel('% peces comerciales')
plt.xticks(range(1914,1924),[str(n) for n in range(1914,1924)])
plt.show()
```
