# Puntos fijos

No todas las ecuaciones tienen puntos fijos que puedan encontrarse algebráicamente. Por ejemplo este sistema:

$$x_{t\text{+1}}=\cos(x_{t})$$

usa una función Tracendental y no puede resolverse con álgebra. 

¿Qué tal graficando?

```python
import numpy as np

# importamos bibliotecas para plotear
import matplotlib
import matplotlib.pyplot as plt

# para desplegar los plots en el notebook
%matplotlib inline
```

```python
def f(x):
    return np.cos(x)
```

```python
x = np.linspace(0,10,100)
f_x = f(x)
fig = plt.plot(x, f_x)
```

```python
def cobweb(f, x0, n, xmin, xmax, ymin, ymax):
        plt.axhline(linewidth=1.0, color="black")
        plt.axvline(linewidth=1.0, color="black")
        plt.ylim((ymin,ymax))
        indep = np.linspace(xmin,xmax,5000)
        diag = lambda x: x
        plt.plot(indep,f(indep),'b')
        plt.plot(indep,diag(indep),'black')
        y0 = f(x0)
        for i in range(n):
                plt.hlines(y0, x0, y0,'r')
                x0 = y0
                y0 = f(x0)
                plt.vlines(x0, x0,y0,'r')


        return
    
cobweb(f,0,10,0,1,0,1)
```
