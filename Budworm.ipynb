```python
# importamos bibliotecas c√≥mputo de matrices
import numpy as np

# importamos bibliotecas para plotear
import matplotlib
import matplotlib.pyplot as plt

# para desplegar los plots en el notebook
%matplotlib inline
```

```python
def predation(B, alfa=50, beta=100):
    return (beta * B**2)/float(alfa**2 + B**2)
```

```python
f0 = plt.plot( [predation(b,5,100) for b in range(100)])
f0 = plt.plot( [predation(b,25,100) for b in range(100)])
f0 = plt.plot( [predation(b,50,100) for b in range(100)])
f0 = plt.plot( [predation(b,55,100) for b in range(100)])
```

```python
def f(B, rB, KB):
    return (rB*B*(1-(B/KB)))-predation(B)

pop=[]
for b in range(30):
    pop.append(f(b,0.5,100))
figura = plt.plot( pop )
```
