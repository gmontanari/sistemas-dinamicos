```python
# importamos bibliotecas c√≥mputo de matrices
import numpy as np

# importamos bibliotecas para plotear
import matplotlib
import matplotlib.pyplot as plt

# para desplegar los plots en el notebook
%matplotlib inline
```

```python
def predation(B, alfa, beta):
    return (beta * B**2)/float(alfa**2 + B**2)
```

```python
pb = []
for b in range(30,300):
    pb.append(predation(b, 50, 100))

```

```python
figura = plt.plot( pb )
```
