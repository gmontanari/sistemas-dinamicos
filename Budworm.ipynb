```python
# importamos bibliotecas c√≥mputo de matrices
import numpy as np

# importamos bibliotecas para plotear
import matplotlib
import matplotlib.pyplot as plt

# para desplegar los plots en el notebook
%matplotlib inline
```

```python
def predation(B, alfa=5, beta=100):
    return (beta * B**2)/float(alfa**2 + B**2)
```

```python
f0 = plt.plot( [predation(b,5,100) for b in range(2,100)])
f0 = plt.plot( [predation(b,25,100) for b in range(2,100)])
f0 = plt.plot( [predation(b,50,100) for b in range(2,100)])
f0 = plt.plot( [predation(b,75,100) for b in range(2,100)])
```

```python
def fprima(B, rB, KB):
    return (rB*B*(1-(B/KB)))-predation(B,0.5,1)

# la poblacion en t=0 es 20
pop=[100,]

t=1
while t<=1001:
    b = pop[t-1]
#    print t,b,fprima(b,0.1,2000),b+fprima(b,0.1,2000)
    pop.append(b+fprima(b,0.03,2000))
    
    t+=1

#pop
figura = plt.plot( pop )
```
