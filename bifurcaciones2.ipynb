```python
from pynamical import logistic_map, simulate, bifurcation_plot

def f(x, l):
    return 4*l*x*(1-x)

pops = simulate(model=f, 
                num_gens=150, 
                rate_min=0, 
                rate_max=1, 
                num_rates=150, jit=False)
bifurcation_plot(pops, xmax=1,
                title="Bifurcaciones",
                xlabel='lambda',
                ylabel='x')
```

```python
from numpy import sin, pi

def g(x, l):
    return l * sin(pi * x)

pop2 = simulate(model=g, 
                num_gens=150, 
                rate_min=0, 
                rate_max=1, 
                num_rates=150, jit=False)
bifurcation_plot(pop2, xmax=1,
                title="Bifurcaciones",
                xlabel='lambda',
                ylabel='x')
```
